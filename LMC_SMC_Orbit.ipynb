Plotting Ekta's Best fit Orbit model for LMC and SMC 

Model : MW1 + LMC2 (1.8e11) 
https://ui.adsabs.harvard.edu/abs/2020ApJ...893..121P/abstract

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all = np.loadtxt('MW1_LMC2_SMC1_orbit_mean_v2.txt') # original B12 Model 1

#columns are: time, MW x, MW y, MW z, MW vx MWvy MWvz etc 
T = all[:,0]
MWx = all[:,1]
MWy = all[:,2]
MWz = all[:,3]
MWvx = all[:,4]
MWvy = all[:,5]
MWvz = all[:,6]
LMCx = all[:,7]
LMCy = all[:,8]
LMCz = all[:,9]
LMCvx = all[:,10]
LMCvy = all[:,11]
LMCvz = all[:,12]
SMCx = all[:,13]
SMCy = all[:,14]
SMCz = all[:,15]
SMCvx = all[:,16]
SMCvy = all[:,17]
SMCvz = all[:,18]
```

```python
# Relative motion of SMC
Sx = SMCx - MWx
Sy = SMCy - MWy
Sz = SMCz - MWz
Svx = SMCvx - MWvx
Svy = SMCvy - MWvy
Svz = SMCvz - MWvz
Sr = np.sqrt(Sx**2 + Sy**2 + Sz**2)
Sv = np.sqrt(Svx**2 + Svy**2 + Svz**2)

# Relative motion of LMC

Lx = LMCx - MWx
Ly = LMCy - MWy
Lz = LMCz - MWz
Lvx = LMCvx - MWvx
Lvy = LMCvy - MWvy
Lvz = LMCvz - MWvz
Lr = np.sqrt(Lx**2 + Ly**2 + Lz**2)
Lv = np.sqrt(Lvx**2 + Lvy**2 + Lvz**2)

# relative 
LSr = np.sqrt( (Lx-Sx)**2 + (Ly-Sy)**2 +(Lz-Sz)**2)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111)

# plot the particle density for M31 
plt.plot(-T, Sr)
plt.plot(-T, Lr)
plt.plot(-T, LSr)

# Add axis labels
plt.xlabel('Look Back Time', fontsize=22)
plt.ylabel('Galactocentric Position (kpc)', fontsize=22)

#set axis limits
#plt.xlim(-20,-160)
#plt.ylim(-250,50)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111)

# plot the particle density for M31 
plt.plot(-T, Sv)
plt.plot(-T, Lv)

# Add axis labels
plt.xlabel('Look Back Time', fontsize=22)
plt.ylabel('Galactocentric Velocity  (kpc)', fontsize=22)

#set axis limits
#plt.xlim(-20,-160)
#plt.ylim(-250,50)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size
```
