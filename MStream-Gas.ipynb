```python
#from plotting import *
#%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import LogNorm
```

```python
# Plot the observational selection function 

#all = np.loadtxt('SMC_StarsPos_Model1.txt')
all = np.loadtxt('StreamGas_Model1.txt')
# x y z 
x = all[:,0] 
y = all[:,1]
z = all[:,2]  
r = np.sqrt(x**2 + y**2 +z**2)

index = np.where((y > 0) & (r<450))
x1 = x[index]
y1 = y[index]
z1 = z[index]
r1 = r[index]

# allv = np.loadtxt('SMC_StarsVel_Model1.txt')
# allv = np.loadtxt('')
# vx vy vz 
#vx = allv[:,0] 
#vy = allv[:,1]
#vz = allv[:,2]  


vx = all[:,3]
vy = all[:,4]
vz = all[:,5]


v =np.sqrt(vx**2 + vy**2 + vz**2) 
v1 = v[index]
```

```python
# For reference,  LMC  COM V and P

Lx = -1.06771
Ly = -41.04967
Lz = -27.83253
Lvx = -57.40
Lvy = -225.59
Lvz = 220.67
      
```

```python
# Radial and Tangential Velocties

vrad = (vx*x + vy*y + vz*z )/r
vtan = np.sqrt(v**2 - vrad**2 )

vrad1 = vrad[index]
vtan1 = vtan[index]
```

```python
# Galactic and Magellanic Stream Coordinates

all = np.loadtxt('StreamGasPos_Model1.txt')
# Galactic l, b
l = all[:,0] 
b = all[:,1]
# Magellanic l, b
ML = all[:,2]  
MB = all[:,3]

```
